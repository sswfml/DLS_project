# Data Exploration: FMA & GTZAN
Explore audio files, metadata, and distribution of genres for the AURA project.

```python
import sys
sys.path.append('..')
from data.dataset import FMADataset, GTZANDataset
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Load FMA metadata
fma = FMADataset('../data/fma_large', chunk_duration=10)
tracks = fma.tracks
tracks.head()
```

```python
# Genre distribution
genre_counts = tracks['track', 'genre_top'].value_counts()
genre_counts.plot(kind='bar', figsize=(12,6))
plt.title('Genre Distribution in FMA Medium')
plt.ylabel('Number of tracks')
plt.show()
```

```python
# Load GTZAN and check genres
gtzan = GTZANDataset('../data/gtzan')
genres = [p.parent.name for p in gtzan.audio_files]
pd.Series(genres).value_counts().plot(kind='bar')
plt.title('GTZAN Genre Distribution')
plt.show()
```

```python
# Audio sample visualization
import librosa.display
sample_audio, sr = librosa.load(gtzan.audio_files[0], sr=16000)
plt.figure(figsize=(12,4))
librosa.display.waveshow(sample_audio, sr=sr)
plt.title('GTZAN sample waveform')
plt.show()
```
