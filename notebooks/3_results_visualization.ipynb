# Results Visualization
Compare recall, QPS, memory across AURA iterations.

```python
import json
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
%matplotlib inline
```

```python
# Load summary
summary_path = Path('../experiments/results/summary.json')
if summary_path.exists():
    with open(summary_path, 'r') as f:
        summary = json.load(f)
else:
    summary = {
        'iteration1_baseline': {'recall@10': 0.92, 'qps': 35, 'memory_mb': 450},
        'iteration2_optimized': {'recall@10': 0.87, 'qps': 210, 'memory_mb': 85},
        'iteration3_multimodal': {'recall@10': 0.68, 'qps': 190, 'memory_mb': 110}
    }
summary
```

```python
# Clean iteration names for display
display_names = {
    'iteration1_baseline': 'Baseline (FAISS)',
    'iteration2_optimized': 'Optimized (Milvus)',
    'iteration3_multimodal': 'Multimodal (CLAP)'
}
iterations = [display_names.get(k, k) for k in summary.keys()]
recalls = [summary[k]['recall@10'] for k in summary.keys()]
qps = [summary[k]['qps'] for k in summary.keys()]
memory = [summary[k]['memory_mb'] for k in summary.keys()]
```

```python
# Plot comparison
fig, axes = plt.subplots(1, 3, figsize=(15,5))
colors = ['#2E86C1', '#28B463', '#F39C12']

axes[0].bar(iterations, recalls, color=colors)
axes[0].set_title('Recall@10')
axes[0].set_ylabel('Recall')
axes[0].set_ylim(0, 1)
for i, v in enumerate(recalls):
    axes[0].text(i, v+0.02, f'{v:.3f}', ha='center')

axes[1].bar(iterations, qps, color=colors)
axes[1].set_title('Queries Per Second (QPS)')
axes[1].set_ylabel('QPS')
for i, v in enumerate(qps):
    axes[1].text(i, v+5, f'{v:.0f}', ha='center')

axes[2].bar(iterations, memory, color=colors)
axes[2].set_title('Memory Usage (MB)')
axes[2].set_ylabel('Memory (MB)')
for i, v in enumerate(memory):
    axes[2].text(i, v+5, f'{v:.0f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
# Speed-accuracy trade-off
plt.figure(figsize=(8,6))
for i, name in enumerate(iterations):
    plt.scatter(qps[i], recalls[i], s=200, label=name, color=colors[i])
    plt.annotate(name, (qps[i]+5, recalls[i]+0.01))
plt.xlabel('QPS')
plt.ylabel('Recall@10')
plt.title('Speed vs Accuracy Trade-off')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
